```python
# FOR LSTM

lstm_data = pd.concat([noaa_atlantic, noaa_pacific], axis=0)
drop = lstm_data.iloc[:, 9:]
lstm_data = lstm_data[~(lstm_data[drop.columns] < -998).any(axis=1)]
new_columns = [
    'ID', 'Name', 'Date', 'Time', 'Latitude', 'Longitude',
    'Maximum Wind', 'Minimum Pressure', 'Low Wind NE', 'Low Wind SE',
    'Low Wind SW', 'Low Wind NW', 'Moderate Wind NE', 'Moderate Wind SE',
    'Moderate Wind SW', 'Moderate Wind NW', 'High Wind NE', 'High Wind SE',
    'High Wind SW', 'High Wind NW'
]
lstm_data = lstm_data[new_columns]

def convert_coordinate(coord):
    if coord[-1] in ['N', 'S']:
        value = float(coord[:-1])
        return value if coord[-1] == 'N' else -value
    elif coord[-1] in ['E', 'W']:
        value = float(coord[:-1])
        return value if coord[-1] == 'E' else -value
    return float(coord)

lstm_data['Latitude'] = lstm_data['Latitude'].apply(convert_coordinate)
lstm_data['Longitude'] = lstm_data['Longitude'].apply(convert_coordinate)
lstm_data['Datetime'] = pd.to_datetime(lstm_data['Date'].astype(str) + lstm_data['Time'].astype(str).str.zfill(4), format='%Y%m%d%H%M')
lstm_data = lstm_data.sort_values(by=['ID', 'Datetime'])
lstm_data.fillna(0, inplace=True)
directory = '/Users/aidanmurphy/Downloads/lstm_data.csv'
lstm_data.to_csv(directory, index=False)
```
